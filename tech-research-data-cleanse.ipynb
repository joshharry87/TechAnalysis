```python
i=0
for item in clean_list:
    clean_list[i] = item.split("                 ")
    i +=1
```

```python
company_names = []
for item in clean_list:
    if len(item) > 1:
        company_names.append(item[1])
    else:
        company_names.append(item[0])
i =0    
for item in company_names:
    company_names[i] = item.strip()
    i +=1
```

```python
core_data = []
i= 0
for item in clean_list:
    index = 0
    core_data.append(item[0])
    current = len(item[0])
    for sub_item in item:
        if len(sub_item) > current:
            core_data[i] = sub_item
            current = len(sub_item)
    i+=1
```

```python
i = 0
for item in core_data:
    core_data[i] = item.split("\n")
    i+=1
```

```python
print(len(item[0]))

item_index = 0
for item in core_data:
    for x  in range(len(item)-1, 0, -1):

        if len(item[x]) == 0:
            del core_data[item_index][x]
    item_index +=1
                
```

```python
split = []

for item in core_data:
    sub_split = []
    for sub_item in item:
        if '%' in sub_item:
            sub_split.append(sub_item)
        
    split.append(sub_split)
    
```

```python
split
index = 0
for item in split:
    organised = []
    for entry in item:
        data = entry.split("%")
       
        organised.append([int(data[0]), data[1]])
    print(organised)
    print(index)
    organised.sort(key= lambda x: x[0])
    organised.reverse()
        
    split[index] = organised
    index +=1
    
```

```python
top_discipline = []
top_percentage = []
second_discipline = []
second_percentage = []
all = []
robotics_percentage = []

index = 0
for item in split:
    print(item)
    if len(item)> 0:
        done = False
        for sub_item in item:
            if not done:
                if 'robot' in sub_item[1].lower():
                    robotics_percentage.append(sub_item[0])
                    done = True
                    
        if not done:
            robotics_percentage.append(0)
        top_discipline.append(item[0][1])
        top_percentage.append(item[0][0])
        second_discipline.append(item[1][1])
        second_percentage.append(item[1][0])
        clean_item = []
        for sub_item in item:
            sub_item[0] = str(sub_item[0])
            data = '% '.join(sub_item)
            clean_item.append(data)
        
        ', '.join(clean_item)
        all.append(clean_item)
    else:
        top_discipline.append('')
        top_percentage.append('')
        second_discipline.append('')
        second_percentage.append('')
        all.append('')
        robotics_percentage.append('')

df['Top Discipline'] = top_discipline
df['top_percentage'] = top_percentage
df['second_discipline'] = second_discipline
df['second_percentage'] = second_percentage
df['robotics_percentage'] = robotics_percentage
df['all'] = all
```
